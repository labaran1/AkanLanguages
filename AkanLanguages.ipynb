```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import os
```

```python
headers = {'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.135 Safari/537.36 Edge/12.246"}
```

```python
def getFanteLinks():
    fanteCategories= ["adwenkyerɛkyerɛ", "agodzisɛm", "amambra", "amanyɛsɛm", "ebibirmu", "eguadzisɛm", "enyigyesɛm", "mpontusɛm", "nwomasua", "dza-ɔforafora"]
    fanteBaseURL='https://akannews.com/fante/category/'
    fanteArticles =[]
    def pushTwiArticle(container):
        for j in range(0, len(container)):
                fanteArticles.append(container[j].a['href'])
    for i in range(0, len(fanteCategories)):
            re = requests.get(fanteBaseURL+fanteCategories[i], headers=headers)
            soup = BeautifulSoup(re.content)
            articlesContainer = soup.findAll('div', class_="elementor-post__card")
            pageNumbers = soup.find_all("a", class_='page-numbers')
            endPage=''
            if(len(pageNumbers) != 0):
                endPage = pageNumbers[-1]["href"].split('/')[-2]
            pushTwiArticle(articlesContainer)
            if(endPage != ''):
                for w in range(2, int(endPage)+1):
                     pageReq = requests.get(fanteBaseURL+fanteCategories[i]+'/page'+str(w), headers=headers)
                     pageSoup = BeautifulSoup(pageReq.content)
                     articlesContainerForPage = pageSoup.findAll('div', class_="elementor-post__card")
                     pushTwiArticle(articlesContainerForPage)        
            else: 
                 continue   
    return fanteArticles
```

```python
def getBlogContent(blogLinks, fileName):
    for i in range(0,len(blogLinks)):
        blog= {}
        print(i)
        req = requests.get(blogLinks[i], headers=headers)
        soup = BeautifulSoup(req.content)
        category = soup.find('div', class_='jeg_meta_category')
        title = soup.find('h1', class_='jeg_post_title')
        blogContent = soup.find('div', class_='content-inner')
        paragraphs = blogContent.findAll('p')
        blog["title"] = title.contents[0]
        blog["category"] =category.span.a.contents[0]
        content = ''
        for j in range(0, len(paragraphs)):
            if paragraphs[j].strong:
                content = content + ' ' + str(paragraphs[j].strong.contents[0])
            elif paragraphs[j].img:
                continue
            else:
                content = content + ' ' + str(paragraphs[j].contents[0]) 
        blog['content'] = content
        sheetName = 'article' + "" + str(i)
        df = pd.Series(blog)
        with pd.ExcelWriter(fileName,
                    mode='a') as writer:  
             df.to_excel(writer, sheet_name=sheetName)
       
   
```
